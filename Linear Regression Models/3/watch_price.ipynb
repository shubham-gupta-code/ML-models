### Model Evaluation

```python
from sklearn.model_selection import cross_val_score
```

```python
lr_model = LinearRegression()

lr_model.fit(X_train, y_train)

y_pred = lr_model.predict(X_test)
r2 = r2_score(y_test, y_pred)

score = cross_val_score(lr_model, X, y, cv=5, scoring='r2')

print(f"Linear Regression r2 score: {r2}")
print('cross_val_score:', score, score.mean(), score.std())
```

```python
dt_model = DecisionTreeRegressor()

dt_model.fit(X_train, y_train)

y_pred = dt_model.predict(X_test)
r2 = r2_score(y_test, y_pred)

score = cross_val_score(dt_model, X, y, cv=5, scoring='r2')

print(f"DecisionTreeRegressor r2 score: {r2}")
print('cross_val_score:', score, score.mean(), score.std())
```

```python
rf_model = RandomForestRegressor()

rf_model.fit(X_train, y_train)

y_pred = rf_model.predict(X_test)
r2 = r2_score(y_test, y_pred)

score = cross_val_score(rf_model, X, y, cv=5, scoring='r2')

print(f"DecisionTreeRegressor r2 score: {r2}")
print('cross_val_score:', score, score.mean(), score.std())
```

```python
xgb_model = xgb.XGBRegressor()

xgb_model.fit(X_train, y_train)

y_pred = xgb_model.predict(X_test)
r2 = r2_score(y_test, y_pred)

score = cross_val_score(xgb_model, X, y, cv=5, scoring='r2')

print(f"DecisionTreeRegressor r2 score: {r2}")
print('cross_val_score:', score, score.mean(), score.std())
```

```python
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train, y_train)

y_pred = ridge_model.predict(X_test)
r2 = r2_score(y_test, y_pred)

score = cross_val_score(ridge_model, X, y, cv=5, scoring='r2')

print(f"Ridge r2 score: {r2}")
print('cross_val_score:', score, score.mean(), score.std())
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'alpha': [0.1, 1.0, 10.0, 100.0] # Alpha is the regularization strength for Ridge regression
}

grid_search = GridSearchCV(estimator=ridge_model, param_grid=param_grid, cv=5, scoring='r2')
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_

best_model = Ridge(**best_params)

best_model.fit(X_train, y_train)
y_pred = best_model.predict(X_test)
r2 = r2_score(y_test, y_pred)

print(f"Ridge R2 score after GridSearchCV: {r2}")
```
